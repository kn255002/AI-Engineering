```python
### Super class for stack...

class Stack():
    '''
    Super class for stack.
    '''
    def __init__(self):
        self.stacklist=[]
        print("Super-calss Instance created to serve Object-id:",id(self))
    def push(self,val):
        self.stacklist.append(val)
    def pull(self):
        var=self.stacklist[-1]
        #self.stacklist.pop()
        del self.stacklist[-1]
        return(var)     
    def show_stack(self):
        print(self.stacklist)

###Child class...

class Addstack(Stack):
    '''
    Child class to use/extend super (Stack) class functionality.
    '''
    def __init__(self):        
        Stack.__init__(self)
        self.__summ = 0
        print("Child-class Instance created to serve Object-id:",id(self))
    def push(self,var):
        self.__summ+=var
        Stack.push(self,var)
    def pull(self):       
        self.__summ-=Stack.pull(self)
    def getsum(self):
        print("Stack summ:",self.__summ)
        return(self.__summ)

    
###
stackobj=Addstack()
print("Object Created:",id(stackobj))
###

stackobj.push(1)
stackobj.push(2)
stackobj.show_stack()
stackobj.getsum()
stackobj.pull()
stackobj.show_stack()
stackobj.__khurram=['k','n']
print(stackobj.__dict__) #List properties (name & value) associated with this object.

```
